```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
#загружаем данные о среднемесячных зарплатах в России
data=pd.read_csv('WAG_C_M.csv',sep=';',index_col=['month'],parse_dates=['month'], dayfirst=True)
```

```python
data.head()
```

```python
%pylab inline
```

```python
#строим график временного ряда
plt.figure(figsize(15,7))
data.plot()
plt.ylabel('Средняя зарплата')
```

```python
from scipy import stats
```

```python
#преобразование Бокса-Кокса для снижения уровня дисперсии
data['disp'],lmbda=stats.boxcox(data.WAG_C_M)
```

```python
data.disp.plot()
```

```python
#сезонное дифференцирование D=1
data['wag_diff']=data.disp-data.disp.shift(12)
plt.figure(figsize(15,7))
data['wag_diff'].plot()
```

```python
import statsmodels.api as sm
```

```python
print('Критерий Дики-Фуллера p=%f'%sm.tsa.stattools.adfuller(data['wag_diff'].dropna().values)[1])
```

```python
#Критерий Дики-Фуллера <0.05 и  позволяет отбросить нулевую гиопотезу о нестационарности ряда.
#Для пущей уверенности проведем обычное дифференцирование
```

```python
#дифференцирование d=1
data['wag_diff2']=data.wag_diff-data.wag_diff.shift(1)
data['wag_diff2'].plot()
```

```python
print('Критерий Дики-Фуллера p=%f'%sm.tsa.stattools.adfuller(data['wag_diff2'].dropna().values)[1])
```

```python
#критерий Дики-Фуллера указывает на стационарность ряда
sm.tsa.seasonal_decompose(data['wag_diff2'][13:]).plot()
plt.show()
```

```python
ax=plt.subplot(211)
sm.graphics.tsa.plot_acf(data.wag_diff2[13:].values.squeeze(),lags=48,ax=ax)
plt.show() #автокорреляция
ax=plt.subplot(212)
sm.graphics.tsa.plot_pacf(data.wag_diff2[13:].values.squeeze(),lags=48,ax=ax)
plt.show() #частичная автокорреляция
```

```python
ps=range(0,2)
d=1
qs=range(0,2)
Ps=range(0,2)
Qs=range(0,1)
```

```python
D=1
```

```python
from itertools import product
```

```python
params=product(ps,qs,Ps,Qs)
```

```python
params_list=list(params)
```

```python
import warnings
```

```python
%%time
results=[]
best_aic=float('inf')
warnings.filterwarnings('ignore')
for param in params_list:
    try:
        model=sm.tsa.statespace.SARIMAX(data.disp, order=(param[0],d,param[1]),
                                        seasonal_order=(param[2],D,param[3],12)).fit(disp=-1)
    except ValueError:
        print('wrong params: ',param)
        continue
    aic=model.aic
    if aic<best_aic:
        best_model=model
        best_aic=aic
        best_param=param
    results.append([param,model.aic])
warnings.filterwarnings('default')
```

```python
result_table=pd.DataFrame(results)
result_table.columns=['parameters','aic']
print(result_table.sort_values(by='aic',ascending=True).head())
```

```python
print(best_model.summary())
```

```python
plt.subplot(211)
best_model.resid[13:].plot()
plt.show()
plt.ylabel('Residuals')
ax=plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[13:].values.squeeze(),lags=48,ax=ax)
plt.show()
```

```python
#по графику нет значимых значений автокорреляции, значит остатки неавтокоррелированы
```

```python
print('КРитерий СТъюдента р=%f'%stats.ttest_1samp(best_model.resid[13:],0)[1])
```

```python
#критерий СТъюдента не отвергает нулевую гипотезу о несмещенности остатков
```

```python
print('КРитерий Дики-Фуллера р=%f'%sm.tsa.stattools.adfuller(best_model.resid[13:].values)[1])
```

```python
#критерий Дики-Фуллера говорит о стационарности остатков
```

```python
print('КРитерий Льюнга-Бокса р=%f'% sm.tsa.stattools.acf(best_model.resid, qstat=True)[2][0]) #свойство resid, хранит остатки модели, qstat=True, означает что применяем указынный тест к коэф-ам
```

```python
#Критерий Льюнга-Бокса не отвергает гипотезу о неавтокоррелированности остатков
```

```python
#обратное преобразование Бокса-Кокса
def invboxcox(y,lmbda):
    if lmbda==0:
        return (np.exp(y))
    else:
        return (np.exp(np.log(lmbda*y+1)/lmbda))
```

```python
data['model']=invboxcox(best_model.fittedvalues,lmbda)
```

```python
plt.figure(figsize(15,15))
data.WAG_C_M.plot()
data.model[13:].plot(color='r')
plt.show()
```
